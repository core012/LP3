```python
n=int(input("Enter no. of items"))
a=float(input("Capacity"))
v=list(map(float, input("Values").split()))
w=list(map(float, input("Weights").split()))
ratio=sorted([(v[i]/w[i], v[i], w[i]) for i in range(n)], reverse = True)
val=0
for r,V,wt in ratio:
    if a>=wt: val+=V; a-=wt
    else: val+=r*a; break
print("Max Value", val)
```
