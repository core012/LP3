```python
def safe(board, r, c, n):
    for i in range(r):
        if board[i][c] == 1 or \
           (c-i+r<n and board[i][c-i+r]==1) or \
           (c+i-r>=0 and board[i][c+i-r]==1):
            return False
    return True

def solve(board, r, n):
    if r == n: return True
    if 1 in board[r]: return solve(board, r+1, n)
    for c in range(n):
        if safe(board, r, c, n):
            board[r][c] = 1
            if solve(board, r+1, n): return True
            board[r][c] = 0
    return False

n = int(input("Enter n: "))
fr, fc = map(int, input("Enter first queen position (row col): ").split())
board = [[0]*n for _ in range(n)]
board[fr][fc] = 1

if solve(board, 0, n):
    for row in board: print(row)
else:
    print("No solution.")
```
