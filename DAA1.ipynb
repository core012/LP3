```python
# Recursive Fibonacci
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n-1) + fib_rec(n-2)

# Print recursive series
def fib_rec_series(n):
    return [fib_rec(i) for i in range(n)]

# Non-recursive Fibonacci
def fib_nonrec_series(n):
    a, b, s = 0, 1, []
    for _ in range(n):
        s.append(a)
        a, b = b, a + b
    return s

n = int(input("Enter number of terms: "))
print("Recursive Series:", fib_rec_series(n))
print("Non-Recursive Series:", fib_nonrec_series(n))
```
